```python
new_text = text[83:]
a =  []
for i in new_text:
    if i == None:
        continue
    a.append(i)
new_text_join = ''.join(a)

b = '\n'
str_list = list(new_text_join)
str_list.insert(38771,b)
str_list.insert(89333,b)
str_list.insert(104735,b)
str_list.insert(165001,b)
str_list.insert(299656,b)
str_list.insert(379302,b)
str_list.insert(405988,b)
str_list.insert(406255,b)
new_text_join = ''.join(str_list)
jie_text = re.split('(第.+节.+\n)',new_text_join)
jie_name = re.findall('(第.+节.+\n)',new_text_join)

disease_text = []
for i in jie_name:
    jie_index = jie_text.index(i)
    if jie_index < len(jie_text):
        jie_describe = jie_text[jie_index+1]
        disease_text.append(jie_describe)
data = pd.DataFrame({'disease':jie_name,'text':disease_text})
data['text_normal'] = ''

reg = "[^0-9A-Za-z\u4e00-\u9fa5]"

fuhao_s =  '【'
fuhao_e = '】'
for i,row in data.iterrows():
    disease_text = row['text']
    text_dict = {}
    text_dict['描述'] = []
    word = re.split(r'(【.+】)',disease_text)    
    for j in word:
        key = re.sub(reg,'',j)
        j_index = word.index(j)
        if j_index < len(word):
            if j_index >= 1:
                word_dict = {}
                if j.startswith(fuhao_s) and j.endswith(fuhao_e):
                    if j in text_dict:
                        word_txt = word[j_index+1]
                        if '（一）' in word_txt:
                            word_txt_content = re.split(r'(（一）.+\n|（二）.+\n|（三）.+\n|（四）.+\n|（五）.+\n|（六）.+\n|（七）.+\n|（八）.+\n|（九）.+\n|（十）.+\n)',word_txt)
                            word_txt_key = re.findall(r'(（一）.+\n|（二）.+\n|（三）.+\n|（四）.+\n|（五）.+\n|（六）.+\n|（七）.+\n|（八）.+\n|（九）.+\n|（十）.+\n)',word_txt)                            
                            for k in word_txt_content:
                                word_txt_index = word_txt_content.index(k)
                                if k in word_txt_key:
                                    word_dict[word_txt_key] = word_txt_content[word_txt_index+1]
                                else:
                                    continue
                            text_dict[key] += word_dict
                        else:
                            text_dict[key] += word[j_index+1]
                                              
                    else:
                        word_txt = word[j_index+1]
                        if '（一）' in word_txt:
                            word_txt_content = re.split(r'(（一）.+\n|（二）.+\n|（三）.+\n|（四）.+\n|（五）.+\n|（六）.+\n|（七）.+\n|（八）.+\n|（九）.+\n|（十）.+\n)',word_txt)
                            word_txt_key = re.findall(r'(（一）.+\n|（二）.+\n|（三）.+\n|（四）.+\n|（五）.+\n|（六）.+\n|（七）.+\n|（八）.+\n|（九）.+\n|（十）.+\n)',word_txt)                            
                            for k in word_txt_key:
                                word_txt_index = word_txt_content.index(k)
                                if k in word_txt_key:
                                    word_dict[k] = word_txt_content[word_txt_index+1]
                                else:
                                    continue
                            text_dict[key] = word_dict
                        else:
                            text_dict[key] = word[j_index+1]                     
                else:
                    continue 
            else:
                text_dict.get('描述').append(word[j_index])
        else:
            text_dict.get('描述').append(word[j_index])
    data.loc[data['disease']==row['disease'],'text_normal'] = [text_dict]
signtext = string.punctuation + '\n'+ '。'+'）'+'“'+'”'+'，'+'O'+'0'
data['disease'] = data['disease'].apply(lambda x:x.translate(str.maketrans('', '', signtext))).apply(lambda x:x.replace(' ','')).apply(lambda x:re.findall('(?<=节).*',x)).apply(lambda x: ",".join(x) if isinstance(x,list) else x)
data.drop(labels=range(54,63),axis=0,inplace = True)
data.drop_duplicates(subset='text',keep='first',inplace=True)
data = data.reset_index(drop=True)
data
```
